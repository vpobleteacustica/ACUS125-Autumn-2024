## 1. Las fuerzas que intervienen y los puntos donde éstas se aplican

> Para analizar <span style="color:red">rotación</span> de un cuerpo debemos considerar las fuerzas que intervienen junto con los puntos donde estas fuerzas se aplica.
> + Cuando estudiamos el problema del péndulo simple, comenzamos haciendo un <span style="color:red">DCL</span> y luego aplicamos la segunda ley de Newton.
> + En este caso del péndulo, consideramos la mas del péndulo como una partícula puntual, en donde la fuerza neta se concentra en un solo punto.
> + Luego, se suman las fuerzas y se obtiene la <span style="color:red">fuerza resultante.</span>

> Pero, las partículas puntuales son una buena aproximación al análisis de un problema real de un cuerpo rígido.
> + En este caso, el cuerpo real podría rotar en torno a sí mismo.
> La rotación respecto a un punto fijo se analiza, por lo tanto, considerando el siguiente par:
> + 1) Los vectores de fuerza que intervienen y
> + 2) Los vectores de posición desde el origen fijo al punto donde se aplica la fuerza.

## 2. Momento de la fuerza

> Este par, da origen a un tercer vector, al que llamaremos <span style="color:red">Momento de la Fuerza $\bf{F}$ respecto al punto $0$:</span>

\begin{align}
{\bf{M}}_{0} &={\bf{r}\times \bf{F}}
\label{eq1}\tag{1}
\end{align}

> De este vector analizaremos matemáticamente su:
> + Dirección.
> + Magnitud.
> + Sentido.

> <span style="color:red">Dirección:</span>
> + $({\bf{r}\times \bf{F}})$ es un vector que es perpendicular al plano formado por $\bf{r}$ y $\bf{F}.$

> <span style="color:red">Magnitud:</span>
> + $|{\bf{r}\times \bf{F}}|$ $=$ $|{\bf{r}}|\cdot |{\bf{F}}|\cdot sin(\theta)$, donde $\theta$ es el ángulo que forman $\bf{r}$ y $\bf{F}.$

> <span style="color:red">Sentido:</span>
> + Usaremos la regla de la mano derecha. El orden de la operación es muy importante: conocer cuál vector está primero: $({\bf{r}\times \bf{F}})$. Así, extendemos la mano derecha y estiramos el pulgar. Los dedos deben seguir al primer vector $({\bf{r}})$ y comenzar a cerrarse. El ángulo $\theta$ comienza siempre desde el primer vector y se dirige hacia el segundo vector $({\bf{F}})$ . 
> + Esa debe ser la posición en que apuntan los dedos.
> + En esta posición, el dedo pulgar de la mano derecha indica tanto la dirección como el sentido de $\bf{M_0}$.

## 3. Momentum angular

> + Este vector es importante para <span style="color:red">evaluar la rotación de un cuerpo</span> respecto a $0$. 
> + Llamamos al vector momentum angular: $\bf{H_0}$.
> + El momentum angular está asociado directamente al vector ${\bf{r}}$, que es el vector de posición del lugar donde se aplica la fuerza ${\bf{F}}$ y también al momentum lineal ${\bf{p}}$.

\begin{align}
{{\bf{H}}}_{0} &={\bf{r}\times \bf{p}}= {\bf{r}}\times m{\bf{\dot{r}}}
\label{eq2}\tag{2}
\end{align}

> + Consideremos la ecuación del momentum angular la que nos va a indicar los factores que hacen cambiar $\bf{H_0}$. Derivamos con respecto al tiempo ${{\bf{H}}_{0}}$ y recordamos que $m$ es constante:

\begin{align}
{\bf{\dot{H}}}_{0} &= {\bf{{\dot{r}}}}\,\times\,m\,{\bf{\dot{r}}} \,+\,{\bf{r}}\,\times\,m\,{\bf{\ddot{r}}}\,=\,{\bf{r}}\,\times\,m\,{\bf{\ddot{r}}}
\label{eq3}\tag{3}
\end{align}

> + Por definición, ${\bf{{\dot{r}}}}\,\times\,m\,{\bf{\dot{r}}} \,=\,{\bf{0}}.$ Recordar que se trata de dos vectores colineales.
> + Pero, por la segunda ley de Newton: $m\,{\bf{\ddot{r}}}={\bf{F}}$, por lo que:

\begin{align}
{\bf{\dot{H}}}_{0} &= \,{\bf{r}}\,\times\,{\bf{F}}
\label{eq4}\tag{4}
\end{align}

> + En conclusión, el factor que hace cambiar $\bf{H_0}$ es el momento de la fuerza:

\begin{align}
{\bf{\dot{H}}}_{0} &= {\bf{M}}_{0}
\label{eq5}\tag{5}
\end{align}


> + Así, los Momentos de fuerza $\bf{M_0}$ son los que pueden cambiar $\bf{H_0}$.
> + $\bf{H_0}$ cambia cuando las fuerzas que intervienen ejercen un <span style="color:red">torque</span> respecto al punto de observación $0$.
> + Cuando ${{\bf{M}}_{0}}={\bf{0}}$, entonces, ${\bf{H}}_{0}={\text{constante}}$, lo que representa el <span style="color:red">principio de conservación del momentum angular</span> es decir, en ausencia de momento de fuerza, el momentum angular es constante.


> + Vale la pena observar aquí que cuando se analiza rotación, ésta no solamente depende del momento lineal $\bf{p}$ sino que también <span style="color:red">depende del punto o lugar</span> respecto del cual observamos la rotación.

> + <span style="color:blue">Análisis 1: Sobre la magnitud de $\bf{M_0}$</span> <span style="color:red">(brazo):</span>
<div>
<img src="brazo2.png" width="500"/>
</div>

> > + $|$$\bf{M_0}$$|$ $=|{\bf{r}\times \bf{F}}|$ $=$ $|{\bf{r}}|\cdot |{\bf{F}}|\cdot sin(\theta)$
> > + Se suele llamar <span style="color:red">brazo de la fuerza</span>, o simplemente <span style="color:red">brazo</span> al término: $|{\bf{r}}|\cdot sin(\theta)$.
> > + Por lo que también se dice que la magnitud de $|$$\bf{M_0}$$|$ es igual a la magnitud de la fuerza $|$$\bf{F}$$|$ multiplicada por su <span style="color:red">brazo</span>. 

> + <span style="color:blue">Análisis 2: Sobre el Momento nulo de la fuerza respecto del origen $0$:</span>

> > + De acuerdo a la definición, ${\bf{M}}_{0}$ $={\bf{r}\times \bf{F}},$ donde ${\bf{r}}$ es el vector de posición del lugar donde se aplica la fuerza ${\bf{F}}$.  
> > + Qué posibilidades hay de que ${\bf{M}}_{0}$ sea ${\bf{0}}$?

> > + Caso 1: Si ${\theta} = 0, \pi$ quiere decir que ${\bf{r}}$ y la fuerza ${\bf{F}}$ son <span style="color:red">colineales</span>. O bien, podemos decir que la fuerza es <span style="color:red">central.</span>
> > + Así, una fuerza central no es capaz de producir un momento de fuerza respecto a $0$.

## Energía del péndulo

> + Ignorando la fricción y cualquier otra fuerza no conservativa, en el péndulo simple la energía total se <span style="color:red">conserva</span>, es decir, 

\begin{align}
\dot{E}&=0 
\label{eq7}\tag{7}
\end{align}

> + Durante el movimiento, la energía total, una combinación de energía potencial gravitacional y energía cinética, consiste en un intercambio constante de energía.

> + La energía <span style="color:red">potencial</span> $V$ del péndulo se puede modelar como:

\begin{align}
V &={\text{m}}{\text{g}}{\text{h}} 
\label{eq8}\tag{8}
\end{align}

> > + con $\text{h}$ la altura como se muestra en la siguiente figura. Cuando $\theta = 0$, el péndulo está en el punto más bajo, $\text{h}=0$, y no existe energía potencial. Mientras que cuando $\theta = \frac{\pi}{2}$, el péndulo está en la parte más elevada, con $\text{h}={\text{L}}$, y está momentáneamente sin movimiento, y toda la energía es potencial. En todos los puntos entre medio, la energía potencial se puede modelar como: $\text{h}={\text{L}}\,(1-\text{cos}(\theta))$. 

\begin{align}
V &={\text{m}}{\text{g}}\,{\text{L}}\,(1-\text{cos}(\theta)) 
\label{eq9}\tag{9}
\end{align}

<p style="text-align:center">
    <a title="Energía potencial"><img src="EnergiaPotencial.png" width="350" height="350" alt="Tiny"></a>

> + La energía cinética de la partícula de masa ${\text{m}}$ se define como:

\begin{align}
T \,&=\,\frac{1}{2}\,{\text{m}}\,{\bf{{\dot{r}}}}\,\cdot{\bf{{\dot{r}}}},\quad {\text{pero,}}\, {\bf{{\dot{r}}}}={\text{L}}\,\dot{\theta}\,\hat{\theta}
\label{eq10}\tag{10}
\end{align}

\begin{align}
T \,&=\,\frac{1}{2}\,{\text{m}}\,{\text{L}}^{2}\dot{\theta}\,^{2}
\label{eq11}\tag{11}
\end{align}

> + Cuando regresa desde $\theta = \frac{\pi}{2}$ al punto cuando $\theta = 0$, el péndulo tiene su mayor velocidad. Toda la energía es cinética y no hay energía potencial.
> + En consecuencia, la energía total es:

\begin{align}
E &= V + T
\label{eq12}\tag{12}
\end{align}

\begin{align}
E &={\text{m}}{\text{g}}{\text{L}}\,(1-\text{cos}(\theta)) + \frac{1}{2}\,{\text{m}}\,{\text{L}}^{2}\dot{\theta}\,^{2}
\label{eq13}\tag{13}
\end{align}

> + Por la ecuación de conservación de la energía,

\begin{align}
\dot{E} &=\,{\text{m}}{\text{g}}{\text{L}}\,{\text{sen}}(\theta)\,\dot{\theta}\,+\,{\text{m}}{\text{L}}^{2}\,\dot{\theta}\,\ddot{\theta}\,=\,0
\label{eq14}\tag{14}
\end{align}

\begin{align}
{\text{m}}{\text{L}}^{2}\,\dot{\theta}\,\ddot{\theta}\,+\,{\text{m}}{\text{g}}{\text{L}}\,{\text{sen}}(\theta)\,\dot{\theta} &=\,0
\label{eq15}\tag{15}
\end{align}

\begin{align}
\dot{\theta}\,({\text{m}}{\text{L}}^{2}\,\ddot{\theta}\,+\,{\text{m}}{\text{g}}{\text{L}}\,{\text{sen}}(\theta)\,)&=\,0
\label{eq16}\tag{16}
\end{align}

> + Existen dos soluciones, ya sea $\dot{\theta}=0$, o bien,

\begin{align}
\ddot{\theta}\,+\,\Bigg(\frac{g}{L}\Bigg)\,{\text{sen}}(\theta)&=0
\label{eq17}\tag{17}
\end{align}

> > + Que es la misma ecuación de movimiento ya obtenida anteriormente. 

> + Analicemos la ecuación de Euler-Lagrange:
> + Para esta ecuación se considera la siguiente combinación de energía potencial y cinética, $\text{V}$ y $\text{T}$, respectivamente:

\begin{align}
\text{L} &=\text{T} - \text{V}
\label{eq18}\tag{18}
\end{align}

> + Aquí, $\text{L}$ se conoce con el nombre de $\bf{Lagrangiana}$. Si ahora se escribe la ecuación:

\begin{align}
\frac{d}{dt}\Bigg( \frac{\partial \text{L}}{\partial \dot{\theta}} \Bigg) -\frac{\partial \text{L}}{\partial \theta}&=0
\label{eq19}\tag{19}
\end{align}

> + donde 

\begin{align}
\text{L} &=\frac{1}{2}\,{\text{m}}\,{\text{L}}^{2}\dot{\theta}\,^{2} - {\text{m}}{\text{g}}{\text{L}}\,(1-\text{cos}(\theta))
\label{eq20}\tag{20}
\end{align}

\begin{align}
\frac{\partial \text{L}}{\partial \theta}&=-\, {\text{m}}{\text{g}}{\text{L}}\,(\text{sen}(\theta))
\label{eq21}\tag{21}
\end{align}

\begin{align}
\frac{\partial \text{L}}{\partial \dot{\theta}} &={\text{m}}\,{\text{L}}^{2}\dot{\theta}
\label{eq22}\tag{22}
\end{align}

> + y derivando con respecto al tiempo:

\begin{align}
\frac{d}{dt}\Bigg( \frac{\partial \text{L}}{\partial \dot{\theta}} \Bigg) &={\text{m}}\,{\text{L}}^{2}\ddot{\theta}
\label{eq23}\tag{23}
\end{align}

> + Usamos la ecuación de Euler-Lagrange y se obtiene:

\begin{align}
\frac{d}{dt}\Bigg( \frac{\partial \text{L}}{\partial \dot{\theta}} \Bigg) -\frac{\partial \text{L}}{\partial \theta}&={\text{m}}\,{\text{L}}^{2}\ddot{\theta}\,+\,{\text{m}}{\text{g}}{\text{L}}\,(\text{sen}(\theta))=0
\label{eq24}\tag{24}
\end{align}

\begin{align}
\ddot{\theta}\,+\,\Bigg(\frac{g}{L}\Bigg)\,{\text{sen}}(\theta)&=0
\label{eq25}\tag{25}
\end{align}
