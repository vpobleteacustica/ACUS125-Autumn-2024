### Módulo de Young:

+ Nos referimos al **módulo de Young**, $E$, también conocido como el módulo de elasticidad, como a la constante de proporcionalidad entre la **Tensión (Esfuerzo)** $\sigma$ y la **Deformación** $\epsilon$, en la zona elástica.

+  El módulo de Young es un indicador de la rigidez del material: si aumentamos $E$ aumentaremos también la rigidez!

+ Se define la Tensión $\sigma$ como la relación cuociente entre la fuerza de deformación $F$ y el área de la sección transversal $A$ del material que se deforma. $F$ es la fuerza deformante significa que actúa perpendicularmente a la sección transversal $A$:

$$
\begin{align}
\sigma &= \frac{F}{A}
\label{eq2}
\end{align}
$$

+ Por otra parte, se define la Deformación $\epsilon$ como el cambio fraccionario de longitud del material cuando este experimenta una Tensión $\sigma$:

$$
\begin{align}
\epsilon &= \frac{\Delta L}{L_0}
\label{eq3}
\end{align}
$$

+  El módulo de Young $E$ es un indicador de la rigidez del material (si aumentamos E aumentaremos también la rigidez).

### Elasticidad de los materiales

+ La **elasticidad** es la tendencia de los materiales sólidos a **volver a su forma original** después de eliminar las **fuerzas externas** que causan una **deformación**. 

* Un material es **elástico** cuando vuelve a su tamaño y forma originales cuando la carga ya no está presente.

* Las razones físicas del comportamiento elástico varían entre los materiales y su comportamiento elástico depende de la estructura microscópica del material: metales, polímeros, caucho. 

### Módulo elástico y límite elástico de un material

+ Estos son dos parámetros que determinan la elasticidad de un material: su **módulo elástico** y su **límite elástico**. 

* Un **módulo elástico alto** es típico de materiales que son muy difíciles de deformar; es decir, materiales que requieren de una gran fuerza para lograr una deformación importante. 

+ Un ejemplo de **módulo elástico alto** es una **barra de acero**. 

+ Un **módulo de elasticidad bajo** es típico de los materiales que se deforman fácilmente bajo una fuerza;

* Un ejemplo de **módulo de elasticidad bajo** es una **barra de goma elástica**. 

+ Si la **Tensión** se vuelve demasiado alta, cuando se retira la fuerza el material ya no vuelve a su forma y tamaño originales, sino que se relaja a una forma y tamaño diferentes: el material se deforma permanentemente.

+ El **límite elástico** es el valor de la **Tensión**, más allá del cual el material **ya no se comporta elásticamente** sino que se **deforma** permanentemente.

+ Un material elástico depende tanto de su **límite elástico** como de su **módulo elástico**. 

+ Un ejemplo, el caucho se caracteriza por un **módulo elástico bajo** y un **límite elástico alto**; por lo tanto, es fácil de estirar y su estiramiento es notablemente grande. 

+ Cuando la fuerza aumenta desde cero, la **Tensión** $\sigma$ resultante es directamente proporcional a la **Deformación** $\epsilon$ en la forma que se indica en la ecuación:

$$
\begin{align}
\sigma &= E\cdot \epsilon
\label{eq1}
\end{align}
$$

+ Esta relación también es conocida como **Ley de Hooke**.

+ Muy similar al caso de un resorte: "$\ldots$ la fuerza restauradora de un resorte es directamente proporcional a un desplazamiento pequeño$\ldots$" En forma de ecuación: 

$$
\begin{align}
F &= k\cdot \Delta x
\end{align}
$$

+ La Ecuación ($\ref{eq1}$) es válida siempre y cuando la **Tensión** $\sigma$ no supere algún valor límite.

+ El límite de proporcionalidad, o **límite de linealidad**, es el valor de Tensión $\sigma$ más grande, más allá del cual $\sigma$ ya no es proporcional a la deformación $\epsilon$. 

+ Más allá del **límite de linealidad**, la relación entre **tensión** y **deformación** ya no es lineal.

+ Cuando la tensión se vuelve mayor que el límite de linealidad pero aún dentro del límite de elasticidad, el comportamiento sigue siendo elástico, pero la relación entre la tensión y la deformación se vuelve no lineal.

+ Para tensiones más allá del límite elástico, un material exhibe un **comportamiento plástico**. 

+ Esto significa que **el material se deforma irreversiblemente** y **no vuelve a su forma y tamaño original**, incluso cuando se retira la fuerza. 

+ Cuando la tensión $\sigma$ aumenta gradualmente más allá del límite elástico, el material sufre una deformación $\epsilon$ **plástica**. 

+ Materiales tipo caucho (gomas) muestran un aumento en $\sigma$ con un incremento en $\epsilon$, lo que significa que este tipo de material llega a ser más difícil de estirar  

+ El valor de la tensión $\sigma$ en el **punto de fractura** se denomina **tensión de rotura** (o tensión última).

### Construir una curva: Tensión $\sigma$ -versus- deformación $\epsilon$

+ Se cuenta con dos bases de datos: una para un material de alumnio 6061 y la otra para un material de acero 1045.

##### 1) Leamos los datos con Pandas y construyamos dos dataframes (df)

```python
steel_df = pd.read_excel('steel1045_v2.xls')
al_df    = pd.read_excel('Al_6061_v2.xls')
```

#### Usemos el método de Panda: df.head() para ver las primeras filas de datos de cada dataframe.

```python
steel_df.head()
```

```python
al_df.head()
```

##### En cada dataframe sólo un par de columnas nos interesan:

##### Las columnas son: FORCE y EXT. 

##### La columna Force contiene los valores de la fuerza aplicada y están en unidades de libras lb. 

##### La columna ext contiene los valores de extensión o deformación, expresados en unidades de %.

##### Un ejemplo, el dato 171 del dataframe: al_df, muestra una fuerza =  9671.362305 lb y una extensión o deformación de 3.003093% lo que es lo mismo que una extensión o deformación de 0.03003093 in/in. 

```python
print(al_df.loc[[170]])
```

##### 2) Usamos estas dos columnas de la siguiente manera: 

+ La columna FORCE la vamos a convertir en tensión $\sigma$, usando la sección de área $A_0$ de las muestras de test empleadas en el dataset. Pare esto, dividimos la fuerza (en lb) por la sección de área $A_0$ de la muestra de prueba.

$$ \sigma = \frac{F}{A_0} $$

+ La sección de área de la muestra $A_0$ es $\pi r^2$. Dado que lo que se mide es el diámetro $d$ de la muestra, y $d$ es dos veces el radio $r$, despejamos $r$ y lo sustituimos:

$$ A_0 = \pi \left(\frac{d}{2}\right)^2 $$

+ Para este ejemplo, tanto para la muestra de prueba de aluminio, como también para la de acero, el diámetro es $d =0.506\,\, (in)$

+ La columna EXT (datos de extesión o deformación mecánica) los convertimos en deformación adimensional (sin unidades):

$$ \epsilon_{adim} = \epsilon_{\%}\times 0.01 \,\,\,\, \left({\text{límite elástico}}\right)$$


```python
d_steel              = 0.506  # unidades en pulgadas: in
stress_steel         = (steel_df['FORCE']*0.001)/(np.pi*((d_steel/2)**2))  # sólo para ajustar la escala a unidad KSI multiplicamos por 0.001
strain_steel_elastic = steel_df['EXT']*0.01  # límite elástico 1% como se muestra en la figura de tensión vs deformación

d_al                 = 0.506
stress_al            = (al_df['FORCE']*0.001)/(np.pi*((d_al/2)**2))           # sólo para ajustar la escala a unidad KSI multiplicamos por 0.001
strain_al_elastic    = al_df['EXT']*0.01     # límite elástico 1% como se muestra en la figura de tensión vs deformación
```

+ En el sistema anglosajón, ksi (kilolibras de fuerza por pulgada cuadrada que se ejercen sobre el material) es una unidad de tensión o de presión.

##### 3) Construimos gráficos: 

```python
fig, ax = plt.subplots(1,1, figsize=(7,5), tight_layout = True)
ax.plot(strain_steel_elastic, stress_steel, strain_al_elastic, stress_al)
ax.set_xlabel('deformación $\epsilon \,(in/in)$')
ax.set_ylabel('tensión $\sigma \,\,(ksi)$')
ax.set_title('diagrama de tensión $\sigma$ versus deformación $\epsilon$ - aluminio y acero')
ax.legend(['acero 1045', 'alumino 6061'])
plt.show()
```

##### Instalar librerías gráficas interactivas bokeh y holoviews:
    + import holoviews as hv
    + hv.extension('bokeh')
    + import bokeh
    
+ Del gráfico de $\sigma$ versus $\epsilon$, podemos hacer zoom en la región lineal.

+ En el caso del alumnio, la región lineal elástica es aproximadamente de 1.5 ksi a 39.5 ksi.

```python
hv.Curve((strain_al_elastic, stress_al),'deformación (in/in)','tensión (ksi)')
```

+ Pandas tiene un método que se llama: 
    
        .between()

+ con el que podemos fijar los límites inferior y superior de la serie de datos de tensión y luego usar esto para indexar tanto las series de datos tensión como deformación en la región lineal elástica: 

```python
E_stress_al = stress_al[stress_al.between(1.5,39.5)]
E_strain_al = strain_al_elastic[stress_al.between(1.5,39.5)]
```

##### Usaremos la función regresión lineal de scipy llamada:
        linregress()

+ para calcular la pendiente en la región lineal elástica.

```python
slope = linregress(E_strain_al, E_stress_al)
print(slope)
```

```python
#print(slope.slope)
#print(slope.intercept)
#print(slope[0])
```

##### Vemos que la pendiente = 9554.61 (ksi). 

##### El valor de r (rvalue) es casi igual a 1 lo que quiere decir que la región elástica lineal es bastante lineal, no perfectamente lineal, pero bastante cercana.

##### Creemos una nueva variable: E_al, para el módulo elástico del alumnio.

```python
E_al_ksi = slope[0]
print('el módulo elástico, en unidades ksi, del aluminio 6061 es:', E_al_ksi)
```

```python
# Expresado en GPa: 1 GPa = 145.03773773 ksi
ksi_ref  = 145.03773773
E_al_GPa = E_al_ksi/ksi_ref
print('el módulo elástico, en unidades MPa, del aluminio 6061 es:', E_al_GPa)
```

```python
# Expresado en N/mm²: 1 N/mm² =  0.14503773772999978 ksi
N_mm2_ref    = 0.14503773772999978
E_al_N_mm2 = E_al_ksi/N_mm2_ref
print('el módulo elástico, en unidades N/mm², del aluminio 6061 es:', E_al_N_mm2)
```

```python
# Expresado en Pa: 1 Pa = 1.4503773773e-07 ksi
Pa_ref = 1.4503773773e-07
E_al_Pa = E_al_ksi/Pa_ref
print('el módulo elástico, en unidades Pa, del aluminio 6061 es:', E_al_Pa)
```
